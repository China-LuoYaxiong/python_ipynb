# 面向过程编程和面向对象编程的思想碰撞

在正式进入话题之前，需要强调的一点是**面向过程编程**和**面向对象编程**与编程语言无关，而是一种编程的思维方式，或者说看待世界的方式。

## 什么是面向过程编程

Procedure Oriented Programming，简记为POP。  
面向过程编程，重点关注的是**过程**二字，什么是**过程**，编程的目的是为了让计算机帮我们解决一些特定的问题，那么大家可以想一下，当你在现实生活中解决某个问题的时候，你一定是一步一步解决问题的，那么这个一步一步的、先干什么再干什么的东西就是**过程**。  
  
基于该思想编写程序就好比在编写一条流水线，是一种机械式的思维方式：  
**优点**：复杂的问题流程化、进而简单化，从而按照流程解决问题即可  
**缺点**：可扩展性差，一条流水线的任意一个环节出现问题都将对后续环节产生影响。 

**总结**：面向过程编程的思想就是，对于一个问题，你的内心**必须知道解决问题的所有步骤**，之后严格按照该步骤，从而达到解决问题的目的。

## 什么是面向对象编程
Object-Oriented Programming，简记为OOP

 ### 为什么会有面向对象编程？

基于前面讲的**面向过程编程**，当你要解决一个问题，你就**必须要知道解决这个问题的所有步骤，之后严格按照该步骤才能解决该问题！**然而，如果你不知道这些所有的步骤呢？一个人的能力是有限的，不可能什么都会。举一个列子：  
你要开一个互联网公司，需要有人干技术吧，需要有人干财务吧，需要有人干业务吧，需要有人干人事吧，作为一个老板，你不可能什么都干，那你会怎么做？你会招会干这些的事的人来做，要招开发人员吧，要招会计吧，招业务人员吧等等，最终推动整个公司向前发展。这些人员都可以称之为一个**对象**，当我们需要解决某个问题的时候，我们并不会自己去做，而是想着，有没有一个对象，或者一些对象共同去解决这个问题，即便实在现实生活中，也是无数不同的对象之间互相交互，才能形成这个缤纷多彩的世界。  
  
那么基于该思想，在写程序的时候，我们要解决某些问题，我们的目的是找到某些对象，让这些对象共同去解决这个问题，如果找不到这些对象，那我们就需要创造一些能够解决这个问题的对象，这就是为什么叫做面向对象编程。

### 那对象又是什么呢？

我们经常听见有人说，Python里面一切皆对象，有字符串对象，有列表对象，有元祖对象等等，那到底什么才是对象呢？**对象就是某些特征与技能的结合体。**特征是什么，比如之前的例子里面，开发人员，会计，业务员，他们都有自己的姓名 年龄 性别等等这些都是他们的特征，同时他们每个都有自己的技能，会计肯定会财务方面的知识，开发人员肯定会编程等等，就像我们常说的专业的事交给专业的人去做，那么这个社会就会非常高效的运转！。

## 类与对象


类，类是什么，是类别，种类？总之就是你认为的那个东西！比如说，老师和大家都是一类，人类，为什么我们是一类？因为我们都会学习吧，会说人话吧等等。那老师和大家也可以属于不同的类，比如大家可以是cda数据分析71期学生类，但老师就不是，这就说明站在不同的维度，类是不同的。**之前说过对象就是某些特征与技能的结合体，那类就是一系列具有相同特征和技能的对象的集合。**这句话我相信大家都能理解，**但是，对于编程而言，这样说或许更贴切：类是一系列对象相同的特征与技能的结合体。因为在定义类的时候，我们定义的是这些特征与技能。通过类实例化出来的对象自动具有类所定义的特征与技能。**

# 类的定义

在编程中时，我们还是用**类是一系列对象相同的特征与技能的结合体，**也就是说我们在定义类的时候其实就是在给类定义特征和技能，为了和其他编程语言概念相似，我们把**特征叫做属性**，把**技能叫做方法。但是，很不幸的是，Python的类里面方法也被称为属性，这是和其他语言不太一样的地方。**

```Python
class 类名:    
    "类的帮助信息"     
    statement            #类体

```
- 类名：一般使用大写字母开头，常使用“驼峰命名法”。
- 类的帮助信息：用于指定类的文档字符串，定义该字符串后。
- 类体：主要由类变量(或类成员)、方法和属性等定义语句组成。可以用pass语句代替。

````python
class Cook:
    '''这是一个厨师的类'''
    
    # 类是一系列对象相同的特征与技能的结合体
    
    # 用变量表示特征（属性）
    name = '张三'
    age  = '27'
    
    # 用函数表示技能（方法）        
    def qie():
        '''这是一个切菜的函数'''
        print('技能：切菜')
    
    def chao():
        '''这是一个炒菜的函数'''
        print('技能：炒菜') 
    
````

这样我们就定义了一个最简单的类，但是明显有一个缺点就是该厨师类当中，所有的厨师的名字和年龄都相同，这个显然有些问题，这里先埋下一个伏笔。

# 定义类中函数必须写上一个参数

报错说：  
这两个函数不需要位置参数，但是呢我们又给了一个？？？我有给吗？？  
肯定还有一些内在的机制在里面我们没有讲到，接下我们再定义一个类看看

````python
class Cook2:
    '''这是一个厨师2的类'''
    
    # 类是一系列对象相同的特征与技能的结合体
    
    # 用变量表示特征（属性）
    name = '张三'
    age  = '27'
    
    # 用函数表示技能（方法）        
    def qie(self):
        print('技能：切菜')
    
    def chao(self):
        print('技能：炒菜')  
````

````python
Cook2.qie(1)
````

````python
ex3 = Cook2()
ex4 = Cook2()
````

````python
ex3.age     # 通过对象调属性这个没有问题
ex4.age
````

````python
ex3.qie()   # 为什么这样就行了呢？
ex4.qie()
````

````python
ex3.chao()  # 为什么这样就行了呢？
ex4.chao()
````

注意观察，类Cook2里面的两个函数qie()和chao()：  
```
class Cook2:
    '''这是一个厨师2的类'''
    
    # 类是一系列对象相同的特征与技能的结合体
    
    # 用变量表示特征（属性）
    name = '张三'
    age  = '27'
    
    # 用函数表示技能（方法）        
    def qie(self):
        print('技能：切菜')
    
    def chao(self):
        print('技能：炒菜')  
```
这两个函数的都有一个参数self,这个不是默认参数吧，因为没有默认值，那调用这个函数的时候就必须给他传入一个实参，这是肯定避免不了的，然而大家再看看，我通过对象调用的时候我并没有给他写入任何实参，那就说明Python的内在机制里面必然有一个自动传参的过程。  
  
比如下面，并没有为函数chao()传入任何实参，但是依然完美执行，说明python内部一定是把某个东西传给了形参self,那到底是把什么传给了self呢？

````python
ex3.chao()  
ex4.chao()
````

````python
Cook.chao
````

````python
ex3.chao    # 通过对象调调的方法叫做绑定方法，什么叫绑定？谁和谁绑定---对象和方法绑定
ex4.chao
````

原因在于，当你要调用这个方法的时候，你得告诉我到底是哪个对象在调方法，所以self接收的其实是对象本身！不信，试试看。

````python
class Cook3:
    '''这是一个厨师3的类'''
    
    # 类是一系列对象相同的特征与技能的结合体
    
    # 用变量表示特征（属性）
    name = '张三'
    age  = '27'
    
    # 用函数表示技能（方法）        
    def qie(self):
        
        print('技能：切菜')
        print('我的self的内容来源于内存地址：', id(self))  # 打印出self的内存地址
    
    def chao(self):
        print('技能：炒菜') 
````

````python
Cook3.qie(1)
````

````python
ex5 = Cook3()   # 实例化两个对象
ex6 = Cook3()
````

````python
ex5.qie()
````

````python
id(ex5)    
````

````python
ex6.qie()
````

````python
id(ex6) 
````

**所以，在类中定义函数的时候，第一个参数必须写上，这个参数用来接收实例对象。**

# 类属性和实例属性

## 类属性

类属性是指定义在类中，但又在函数体外的属性。类属性可以被类的所有实例所共享。  
对于下面这个类的两个属性name和age就是类属性，对于该类实例化的对象都是一样的,但是往往是我们需要实例化出属性不同的对象。

````python
class Cook3:
    '''这是一个厨师3的类'''
    
    # 类是一系列对象相同的特征与技能的结合体
    
    # 类属性
    name = '张三'        
    age  = '27'
    
    # 用函数表示技能（方法）        
    def qie(self):          # 类中定义的函数必须有一个参数接收实例对象
        print('技能：切菜')
       
    def chao(self):
        print('技能：炒菜') 
````

这个时候就必须要有一个初始化的过程,需要一个特殊的函数 ```__init__()  ```来帮我们完成这样一个功能,类似Java里面这个叫做构造函数。**每当创建一个新的实例的时候，python会自动执行这个函数**。和其他函数一样，这个构造函数也必须有一个参数接收实例对象本身，并且必须是第一个参数。

````python
class Cook4:
    '''这是一个厨师4的类'''
    
    # 类是一系列对象相同的特征与技能的结合体
    
    # 用变量表示特征（属性）
    name = '张三'        
    age  = '27'
    
    # 特殊函数，__init__()
    def __init__(self): # 类中定义的函数必须有一个参数接收实例对象，__init__()也是
        print('__init()__在实例化对象的时候就会被执行')
    
    # 用函数表示技能（方法）        
    def qie(self): 
        print('技能：切菜')
       
    def chao(self):
        print('技能：炒菜')   
        
    
    
````

````python
ex7 = Cook4()   # __init__()实例化的时候会被执行，其他函数不会
````

````python
ex8 = Cook4()  # 每次实例化的时候都会被执行
````

## 实例属性

我们需要把实例化不同属性的对象时，可以把这些需要改变的属性定义在```__init__()```函数内部，这种属性可以称之为实例属性。当然需要你把指定值传给这些属性，就像函数一样。

````python
class Cook5:
    '''这是一个厨师4的类'''
    # 类是一系列对象相同的特征与技能的结合体
    
    # 类属性
    country = '中国'
    
    
    # 特殊函数，__init__()
                  #ex9  '张三' 26
    def __init__(self, name, age): 
        # 实例属性 

        
        self.name = name               
        self.age  = age
        print('__init()__在实例化对象的时候就会被执行')
    
    # 用函数表示技能（方法）        
    def qie(self): 
        print('技能：切菜')
       
    def chao(self):
        print('技能：炒菜')         
````

````python
ex9 = Cook5('张三', 26)
````

````python
ex9.country
ex9.name
ex9.age
````

````python
ex9.qie()
````

````python
ex9.chao()
````

````python
ex10 = Cook5('李四', 32)
````

````python
ex9.名字    # 实例属性
ex10.名字
````

````python
ex9.age    # 实例属性
ex10.age
````

````python
ex9.country  # 类属性
ex10.country
````

# 练习 
定义一个计算器Calculator的类

**类属性**：  
产地（Place_of_Origin） = ‘made in China'  
  
**实例属性**：  
颜色（colour）  
品牌（brand）  
价格（price）

**方法（技能）**：  
sum(a, b)求两个数的和

summ(a, b, c, d,....)求一系列数的和

summ1(a, b, c, d,...)求一系列数的倒数和  
    
**最后实例化一个对象出来试试功能**


````python
 class Calculator:
        '''这是一个计算器'''
        
        # 类属性
        产地 = 'made in China'
        
        # 实例属性
        def __init__(self, a, b, c):
            '''这是初始化函数'''
            self.颜色 = a
            self.品牌 = b
            self.价格 = c
            
        # 方法
        def sum(self, a, b):
            '''这是一个求两个数的和的函数'''
            print(a+b)
        
        def summ(self, *a):
            '''这是一个求一系列数的和的函数'''
            s = 0
            for i in a:
                s += i
            print(s)
        
        def summ1(self, *a):
            '''这是一个求一系列数的倒数和的函数'''
            s = 0
            if 0 in a:
                print("不能含有0, 请重新输入！")
            else:
                for i in a:
                    s += 1/i  
            print(s)  
            
        
````

````python
x1 = Calculator('黑色',  '淘宝', 100)
````

````python
x1.产地
x1.价格
x1.品牌
x1.颜色
````

````python
x1.sum(1, 3)
````

````python
x1.summ(1, 2, 3, 4, 4)
````

````python
x1.summ1(1, 2, 4)
````

# 动态的往类中添加方法  
比如刚才我们定义的计算器的类中，想再添加一个功能，  
summ2(a, b, c, d, ...)求一系列数平方和的函数  
除了在类的定义中添加以外，还可通过types库的中的“MethodType”将自定义函数添加到类中  
**添加后，原来的实例也会有这种新增加的方法**

````python
def summ2(self, *args):
    '''这是一个求一系列数平方和的函数'''
    s = 0
    for i in args:
        s += i**2
    print(s)  
````

````python
summ2(1, 2, 3)   # 为什么是13？
````

````python
from types import MethodType
````

````python
Calculator.summ2 = MethodType(summ2,Calculator)
````

````python
x = Calculator('黑色', '淘宝', '28元')
````

````python
x.summ2.__doc__
````

````python
x.summ2(1, 2, 3)
````

## 练习  
往刚才的计算器Calculator的类里，添加一个功能  
summ3(a, b, c, d, ...)求出一组数字里第二大和第二小的差的平方


````python
def summ3(*args):
    list_args = list(args)  # 将元祖转化为列表
    print(list_args)    
    list_args.sort()    # 默认进行升序
    print(list_args)
    y = (list_args[-2]-list_args[1])**2
    print(y)
````

````python
summ3(2, 4, 6, 8, 1, 5, 7 )
````

````python
def summ3(self, *args):
    ''')求出一组数字里第二大和第二小的差的平方'''
    list_args = list(args)  # 将元祖转化为列表
    print(list_args)    
    list_args.sort()    # 默认进行升序
    print(list_args)
    y = (list_args[-2]-list_args[1])**2
    print(y)
````

````python
from types import MethodType
Calculator.summ3 = MethodType(summ3,Calculator)
````

````python
x2 = Calculator('黑色', '淘宝', '28元')
````

````python
x2.summ3(1, 2, 3, 6, 9)
````
