### while循环

只要while后面的条件判断为True，那么就会一直执行循环体内的语句。

```python
while 条件:
        条件满足时，做的事情1
        条件满足时，做的事情2
        条件满足时，做的事情3
        ...(省略)...
```

**例1：打印出1-100**

````python
i = 1
while i<=100:
    print('python')
    i += 1
````

#### while循环嵌套（难点）

使用while循环嵌套打印出九九乘法表：


```python
1*1=1	
1*2=2	2*2=4	
1*3=3	2*3=6	3*3=9	
1*4=4	2*4=8	3*4=12	4*4=16	
1*5=5	2*5=10	3*5=15	4*5=20	5*5=25	
1*6=6	2*6=12	3*6=18	4*6=24	5*6=30	6*6=36	
1*7=7	2*7=14	3*7=21	4*7=28	5*7=35	6*7=42	7*7=49	
1*8=8	2*8=16	3*8=24	4*8=32	5*8=40	6*8=48	7*8=56	8*8=64	
1*9=9	2*9=18	3*9=27	4*9=36	5*9=45	6*9=54	7*9=63	8*9=72	9*9=81 
```

````python
i = 1          
while i <= 9:  #控制行的循环
    j = 1
    while j <=i:   # 控制列的循环
        
        print("{}*{}={}".format(j,i,j*i),end='\t')
        j += 1
    i += 1
    print('')
````

## 跳转语句

### break语句

&emsp;&emsp;break语句可以终止当前的循环，包括while和for在内的所有控制语句。
<br/>&emsp;&emsp;这就好比你绕圈跑步，一开始打算跑5圈。</br>
<br/>&emsp;&emsp;跑了2圈半，感觉要拉肚子了，你就停下来，不跑了，找厕所去了。  
——结束当前循环</br>
<br/>&emsp;&emsp;此时，算作你跑了2圈。</br>


#### 在单层的for循环中

````python
for i in '123456789':
    print(i)
    
    
print('我是在循环结束之后再输出！')
````

**使用break语句， 当i=‘5’的时候就打断整个循环。**

##### 练习

**找出前100个能被3整除的正整数**

#### 在多层的for循环当中（选讲）

````python
name = '世界杯在召唤我'

for i in range(3):
    for x in name:
        if x == '在': 
             break       #打断的是最内层的for循环，但仍然继续执行外层循环
        print(x)
    print("——————我是{}次大循环结束之后的优美的分割线——————".format(i+1))

print("这句话在for循环后，与for循环无关")
````

**注意多层循环的时候，打断的是最内层的循环**

&emsp;  

&emsp;  
#### 在while循环中

````python
i = 0
while i<10:
    if i==5:
        break
    print(i)
    i += 1
    
    
print('while循环结束之后我才输出')
````

**多层的while循环也是打断的是最内层的循环**

### continue
- continue的作用：用来结束本次循环，紧接着执行下一次的循环。
- 这就好比你绕圈跑步，一开始打算跑5圈。 
- 跑了2圈半，发现好像落下钱包在起跑点了，终止跑第二圈。
- 然后你回到起跑点找钱包，找到钱包，然后继续重新跑，算作从第3圈开始跑。
- 这里你要注意的是：<font color=#DC143C>
    - break用来结束当前代码块的最内层循环。
    - continue用来结束当前代码块的本次循环。

#### 在for循环中

````python
name = '世界杯在召唤我'

for x in name:
    if x == '在': 
        continue
    print(x)
````

&emsp;  

&emsp;  
#### 在while循环中

````python
i = 0
while i<6:
    i = i+1
    if i==3:
        continue
    print(i)
````

<font color=#DC143C size=4 face="黑体">  
**注意:**
    
<font color=#00008B size=3 face="黑体">  
**break/continue只能用在循环中，除此以外不能单独使用**   

**break/continue在嵌套循环中，只对最近的一层循环起作用**

### pass空语句

````python
for i in 'Python':
    if i == 'h':
        pass
    else:
        print(i)
````

````python
for i in 'Python':
    if i == 'h':
       # pass      # 这样会使程序不完整，会报错
    else:
        print(i)
````

**如果我们上面的pass不写（或者用注释代替），会报错，因为if语句是不完整的：**
```
从上面对比可以知道，pass是空语句，是为了保持程序结构的完整性。
pass 不做任何事情，一般用做占位语句。
```
