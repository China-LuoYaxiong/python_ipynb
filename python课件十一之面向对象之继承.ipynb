# 继承

## 什么是继承？

**面向对象编程的三大特性：封装 继承 多态**

举一个列子，矩形 菱形 平行四边形和梯形都是四边形。因为四边形具有一个共同的特征：拥有四条边。只要将一个四边形的四条边继续做一些调整，就会得到上述图形。以平行四边形为例，如果把平行四边形看作四边形的延伸，那么平行四边形就复用了四边形的特征和技能，但同时也有自己特有的属性和技能，比如，平行四边形的对边平行且相等。在Python中，可以把**平行四边形类看作是继承四边形类后产生的类**，其中将类似于平行四边形的类称为**子类或者派生类**，将类似于四边形的类称为**父类**或者**超类**或者**基类**。值得注意的是，在阐述平行四边形与四边形的关系时，我们可以说平行四边形是特殊的四边形，但是不能反过来说四边形是平行四边形。同理，在Python中，子类的实例都是父类的实例，但不能说父类的实例是子类的实例。  
  
**总之，继承本质上就是类与类之间的一种关系，子类能够继承父类拥有的特征和技能，从而增加代码的重用。**

## Python中继承的基本语法

```
class Classname(baseclasslist):
    '''类的说明文档'''
    statement           # 类体

```
Classname:类名  
baseclasslist：用于指定要继承的基类，可以有多个，类名之间用逗号分隔，如果不指定，将默认使用所有Python对象的根类object。

````python
# 定义一个类A
class A1:
    '''   '''
    def a(self):
        print('hello, 我来自于A1')    
````

````python
# 定义一个类B
class B1:
    '''   '''
    def b(self):
        print('hello, 我来自于B1')    
````

````python
# 定义一个类C, 父类是A1 和 B1
class C1(A1, B1):
    def c(self):
        print('hello, 我来自于C1') 
````

**C1类将直接继承其父类的所有属性和方法**

````python
C1.c(1)
````

**既然C1类继承了其父类的属性和方法，则其所实例化的对象也自动继承其父类A 和 B所拥有的属性和方法**

````python
exc1 = C1()
````

````python
exc1.a()
````

````python
exc1.b()
````

````python
exc1.c()
````

### 首先定义一个类：小学生计算器
```
      类属性： 
             产地=中国

       方法：  
            打印两个数的和
            打印两个数的差
            打印两个数的乘
            打印两个数的商
             
```
    
      


````python
class 小学生计算器:
    '''这是一个小学生计算器'''
    
    # 类属性
    产地 = '中国'
    
    # 方法
    def 打印两个数的和(self, a, b):
        '''这是一个打印两个数的和的函数'''
        print(a+b)
    
    def 打印两个数的差(self, a, b):
        '''这是一个打印两个数的差的函数'''
        print(a-b)
    
    def 打印两个数的乘(self, a, b):
        '''这是一个打印两个数的乘的函数'''
        print(a*b)
     
    def  打印两个数的商(self, a, b):
        '''这是一个打印两个数的商的函数'''
        if b == 0:
            print("分母不能为0！")
        else:
            print(a/b)
````

````python
x1 = 小学生计算器()
````

````python
x1.产地
````

### 在定义一个：中学生计算器，使其继承小学生计算器  
```
并且新增三个方法：
            求e得x次方
            求sigmoid函数
            求tanh函数
```

$$
sigmoid:\frac{1}{1+e^{-x}}
$$

$$
tanh:\frac{e^{x}-e^{-x}}{e^{x}+e^{-x}}
$$

````python
import math
````

````python
class 中学生计算器(小学生计算器):
    '''这是一个中学生计算器'''
    # 方法
    def 求e得x次方(self, x):
        '''这是一个求e得x次方的函数'''
        print(math.exp(x))
        
    def 求sigmoid函数(self, x):
        '''这是求sigmoid函数'''
        print(1/(1+math.exp(-x)))
        
    def 求tanh函数(self, x):
        '''这是一个求tanh的函数'''
        print(math.tanh(x))    
````

## 方法重写
**当父类的某个方法不完全适用于子类时，可以在子类里重新改写这个方法。此时子类和父类中都有这个函数，将优先使用子类中的**

````python
class A2:
    '''   '''
    def a(self):
        print('hello, 我来自于A2') 
````

````python
class C2(A2):
    def a(self):
        print('hello, 我来自于C2') 
````

````python
exc2 = C2()
````

````python
exc2.a()  # 优先使用子类自己的方法，如果没有这个方法才会去父类里面找
````

### 方式一：指名道姓地调用（其实与继承没有什么关系的）

````python
CdaPeople.__init__   # 这就是一个函数
````

````python
class CdaTeacher_1:
    
    '''这是一个cda老师的类'''
    
    # 类属性
    country  = '中国'
    
    # 实例属性   exct1 '王老师', 30, '男', 3500, 'T4'
    def __init__(self, name, age, sex, salary, level):
#         self.name = name
#         self.age  = age
#         self.sex  = sex
#       以上三行代码就可以用下面这句代码替换
#       本质上就是调用类CdaPeople中的__init__()函数而已
                           
#                          exct1 '王老师', 30, '男',
        CdaPeople.__init__(self, name, age, sex)  #调用函数
        
        self.salary = salary
        self.level  = level
        
    # 技能（方法）
    def learn(self):
        print('技能：学习')
    
    def teaching(self):
        print('技能：授课')
````

### 方式二：super()调用（严格依赖于继承）

**super()的返回值是一个特殊的对象，该对象专门用来调用父类中的属性**

````python
class CdaTeacher_2(CdaPeople):
    
    '''这是一个cda老师的类'''
    
    # 类属性
    country  = '中国'
    
    # 实例属性   exct2 '王老师', 30, '男', 3500, 'T4'
    def __init__(self, name, age, sex, salary, level):
#         self.name = name
#         self.age  = age
#         self.sex  = sex
#       以上三行代码就可以用下面这句代码替换
#       super()对象直接调用父类的__init__()方法
        super().__init__(self, name, age, sex)
        
        self.salary = salary
        self.level  = level
        
    # 技能（方法）
    def learn(self):
        print('技能：学习')
    
    def teaching(self):
        print('技能：授课')
````

````python
exct2 = CdaTeacher_2('王老师', 30, '男', 3500, 'T4')  # 为什么报错了？？？
````

**super()是一个特殊对象，只要是对象，通过对象调的方法就像之前讲的叫绑定方法，会有一个自动传参的过程。  
当通过supper()调用父类中__init__()函数的时候，第一个参数self不必传入实参**

````python
class CdaTeacher_3(CdaPeople):
    
    '''这是一个cda老师的类'''

    
    # 实例属性   exct3 '王老师', 30, '男', 3500, 'T4'
    def __init__(self, name, age, sex, salary, level):
#         self.name = name
#         self.age  = age
#         self.sex  = sex
#       以上三行代码就可以用下面这句代码替换
#       super()对象直接调用父类的__init__()方法
        super().__init__(name, age, sex)     
        
        self.salary = salary
        self.level  = level
    
    def teaching(self):
        print('技能：授课')
````

````python
exct3 = CdaTeacher_3('王老师', 30, '男', 3500, 'T4')
````

````python
exct3.country
exct3.name
exct3.age
exct3.sex
exct3.salary
exct3.level
````

````python
exct3.learn()
````

````python
exct3.teaching()
````

# 模拟男女谈恋爱

## 定义一个男朋友类
```
类属性：
     性别=男
     
实例属性：
     姓名
     年龄
     金钱值（某种程度上代表身上的余额）范围（0-100）
     开心值（某种程度上代表其愉悦程度）范围（0-100）
     
方法：
     抽烟：此方法会使开心值+3，金钱值-1
     喝酒：此方法会使开心值+1
```
     
     
     


````python
class 男朋友:
    '''这是一个男朋友的类'''
    
    # 类属性
    性别 = '男'
    
    # 实例属性
    def __init__(self, name, age, m, h):
        self.姓名   = name
        self.年龄   = age
        self.金钱值 = m
        self.开心值 = h
        
    def 抽烟(self):
        self.开心值 += 3
        self.金钱值 -= 1
        
    def 喝酒(self):
        self.开心值 += 1      
        
    def 打印所有信息(self):
        print(self.姓名, self.年龄, self.金钱值, self.开心值)
````

````python
小明 = 男朋友('小明', 24, 68, 89)
````

````python
小明.打印所有信息()
````

````python
小明.抽烟()
````

````python
小明.打印所有信息()
````

````python
小明.喝酒()
````

````python
小明.打印所有信息()
````

## 定义一个女朋友类  
```
类属性：
     性别=女
     
实例属性：
     姓名
     年龄
     金钱值（某种程度上代表身上的余额）范围（0-100）
     开心值（某种程度上代表其愉悦程度）范围（0-100）
     
方法：
     购物：此方法会使开心值+3，金钱值-1
     逛街：此方法会使开心值+2
```

````python
class 女朋友:
    '''这是一个女朋友的类'''
    
    # 类属性
    性别 = '女'
    
    # 实例属性
    def __init__(self, name, age, m, h):
        self.姓名   = name
        self.年龄   = age
        self.金钱值 = m
        self.开心值 = h
        
    def 购物(self, x):
        self.开心值 += 3
        self.金钱值 -= 1
        x.金钱值 -= 5
        
    def 逛街(self):
        self.开心值 += 2    
        
    def 打印所有信息(self):
        print(self.姓名, self.年龄, self.金钱值, self.开心值)     
````

````python
小婉 = 女朋友('小婉',  23, 78, 68)
````

````python
小婉.打印所有信息()
````

````python
小婉.购物()
````

````python
小婉.打印所有信息()
````
